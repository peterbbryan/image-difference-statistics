## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from image_difference_statistics import similarity_metrics
from image_difference_statistics.data import camera
from image_difference_statistics.plots import imshow, imshow_many
```

## Configure

```python
sns.set_theme()
```

## Original

```python
imdata = camera()
imshow(imdata=imdata)
```

## Rotations

```python
from image_difference_statistics.transforms.euclidean import rotate

angles = np.linspace(0, 360, 10)
rotated_imdata_list = [rotate(imdata=imdata, angle=angle) for angle in angles]

imshow_many(imdata_list=rotated_imdata_list)
```

### SSIM

```python
ssims = [similarity_metrics.ssim(imdata, rotated_imdata) for rotated_imdata in rotated_imdata_list]
plt.plot(ssims)
plt.title("SSIM calculated for rotated samples")
plt.xlabel("Image index")
plt.xticks(np.arange(len(ssims)))
plt.ylabel("SSIM")
plt.show()
```

```python
mutual_information = [similarity_metrics.normalized_mutual_information(imdata, rotated_imdata) for rotated_imdata in rotated_imdata_list]
plt.plot(mutual_information)
plt.title("Mutual information calculated for rotated samples")
plt.xlabel("Image index")
plt.xticks(np.arange(len(ssims)))
plt.ylabel("Mutual information")
plt.show()
```
